# 1. 데이터 형태 파악

```python
import pandas as pd

# CSV 파일로부터 데이터 불러오기
# data = pd.read_csv('[24-2 DS_Project1] Data_Submission/[24-2 DS_Project1] Data.csv') -> 원본파일 이름
data = pd.read_csv('/Users/gidaseul/Documents/GitHub/DS/datas/train.csv')

# 데이터프레임의 기본 정보 확인
print("데이터 정보:\n", data.info())  # 각 열의 타입, 결측치 등을 확인

# 데이터의 첫 몇 줄 확인
print("데이터 첫 5줄:\n", data.head())

# 데이터의 컬럼명 확인
print("컬럼명:\n", data.columns)

# 결측치 확인
print("결측치:\n", data.isnull().sum())

# 기술 통계 확인
print("기술 통계:\n", data.describe())
```

```python
# 전체 컬럼의 고유값과 고유값의 개수 확인
for column in data.columns:
    unique_values = data[column].unique()
    unique_count = len(unique_values)
    print(f"Column '{column}' has {unique_count} unique values:")
    print(unique_values)
    print("-" * 50)
```

```python

# 범주형 변수 라벨링
data['Has Chip'] = data['Has Chip'].map({True: 1, False: 0})
data['Is Fraud?'] = data['Is Fraud?'].map({'Yes': 1, 'No': 0})

# 범주형 변수 라벨링 추가
data['Card Brand'] = data['Card Brand'].map({'Visa': 0, 'Mastercard': 1, 'Others': 2})
data['Card Type'] = data['Card Type'].map({'Debit': 0, 'Credit': 1, 'Debit (Prepaid)': 2})

# 'Error Message'의 NaN 값을 0으로, 나머지 값을 1번부터 라벨링
data['Error Message'] = data['Error Message'].fillna('No Error')
data['Error Message'], error_message_unique = pd.factorize(data['Error Message'])
data['Error Message'] = data['Error Message'].map(lambda x: 0 if error_message_unique[x] == 'No Error' else x + 1)

# 변환된 데이터 확인
print(data.head())

# 라벨링 표 생성
label_mapping = {
    'Has Chip': {1: 'True', 0: 'False'},
    'Is Fraud?': {1: 'Yes', 0: 'No'},
    'Card Brand': {0: 'Visa', 1: 'Mastercard', 2: 'Others'},
    'Card Type': {0: 'Debit', 1: 'Credit', 2: 'Debit (Prepaid)'},
    'Error Message': {i + 1 if value != 'No Error' else 0: value for i, value in enumerate(error_message_unique)}
}

# 라벨링 표 출력
for column, mapping in label_mapping.items():
    print(f"Column '{column}' Label Mapping:")
    for key, value in mapping.items():
        print(f"{key}: '{value}'")
    print("-" * 50)
```

# 날짜 데이터 관련 처리 -> 날짜 형식을 날짜 데이터에 맞게 처리함

```python
# day=1을 기본으로 설정하여 연월 데이터만 있는 경우에도 datetime으로 변환
df=data
# Expires와 Acct Open Date 컬럼을 datetime 형식으로 변환 (MM/YYYY 형식)
# 변환 후 YYYY-MM 형식으로 포맷팅
df['Expires'] = pd.to_datetime(df['Expires'], format='%m/%Y', errors='coerce').dt.strftime('%Y-%m')
df['Acct Open Date'] = pd.to_datetime(df['Acct Open Date'], format='%m/%Y', errors='coerce').dt.strftime('%Y-%m')

# Year, Month, Day 컬럼을 datetime 형식으로 변환하여 Transaction Date 생성 카드 거래 날짜로 만들어서 사용.
df['Transaction Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']].astype(str).agg('-'.join, axis=1), errors='coerce')

# Year PIN last Changed 컬럼을 datetime 형식으로 변환하고, 연도만 필요하므로 day와 month를 기본 값으로 설정
df['Year PIN last Changed'] = pd.to_datetime(df['Year PIN last Changed'].astype(str) + '-01-01', errors='coerce').dt.date

# 변환된 데이터프레임을 확인
print(df[['Expires', 'Acct Open Date', 'Transaction Date', 'Year PIN last Changed']].head())
```

# 내가 생각하기에는 각 유저가 사기를 당한 시점의 특성들 중에서 공통적으로 엮이는 특성이 있는지 그리고 그것이 사기가 안 당한 사람한테 실제로 반대로 작용했는지가 보고 싶고, 게다가 클래스가 불균형하다는 문제점이 있기 때문에 

```python
# 유저별 사기 거래 횟수 및 비율 계산
user_fraud_counts = df.groupby('User')['Is Fraud?'].value_counts(normalize=True).unstack(fill_value=0)
user_fraud_counts

```

```python
# 각 유저별로 사기 거래가 발생한 시점 및 거래 특성 확인
fraud_transactions = df[df['Is Fraud?'] == 1].sort_values(by=['User', 'Transaction Date'])
fraud_transactions
```

```python
# 유저별 사기 거래 비율 확인
user_fraud_ratio = fraud_transactions.groupby('User')['Is Fraud?'].count() / df.groupby('User')['Is Fraud?'].count()
user_fraud_ratio
```
